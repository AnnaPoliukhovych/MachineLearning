8. Створіть матрицю розміром 2x2, заповніть її випадковими цілими числами в діапазоні від 1 до 12 та знайдіть її обернену матрицю.

```python
import numpy as np

# Створюємо матрицю 2x2 з випадковими числами від 1 до 12
matrix = np.random.randint(1, 13, size=(2, 2))

print("Матриця 2x2:")
print(matrix)

# Перевіряємо, чи матриця обернена (детермінант не нуль)
det = np.linalg.det(matrix) # обчислює детермінант
if det != 0:
    inverse_matrix = np.linalg.inv(matrix) # знаходить обернену матрицю
    print("\nОбернена матриця:")
    print(inverse_matrix)
else:
    print("\nМатриця не обернена (детермінант = 0).")
```

```python
import numpy as np

# Створюємо матрицю 3x3 з випадковими дійсними числами від 0 до 1
matrix = np.random.rand(3, 3)

print("Матриця 3x3:")
print(matrix)

# Транспонуємо матрицю
transposed_matrix = matrix.T

print("\nТранспонована матриця:")
print(transposed_matrix)
```

13. Створіть дві матриці розміром 4x4, заповніть їх випадковими дійсними числами в
діапазоні від -2 до 2 та знайдіть добуток їх елементів.

```python
import numpy as np

# Створюємо дві матриці 4x4 з випадковими дійсними числами від -2 до 2
matrix1 = np.random.uniform(-2, 2, size=(4, 4))
matrix2 = np.random.uniform(-2, 2, size=(4, 4))

print("Матриця 1:")
print(matrix1)

print("\nМатриця 2:")
print(matrix2)

# Поелементно множимо матриці
elementwise_product = matrix1 * matrix2
print("\nПоелементний добуток матриць:")
print(elementwise_product)

# Добуток усіх елементів результату
total_product = np.prod(elementwise_product)
print("\nДобуток всіх елементів:")
print(total_product)
```

15. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в
діапазоні від 0 до 1 та знайдіть вектор-стовпчик, що містить суму елементів кожного
стовпця матриці.

```python
import numpy as np

# Створюємо матрицю 3x3 з випадковими дійсними числами від 0 до 1
matrix = np.random.rand(3, 3)

print("Матриця 3x3:")
print(matrix)

# Знаходимо суму елементів по стовпцях
column_sums = np.sum(matrix, axis=0)

# Перетворюємо вектор рядок у вектор-стовпчик
column_sums_column_vector = column_sums.reshape(-1, 1)

print("\nВектор-стовпчик сум елементів кожного стовпця:")
print(column_sums_column_vector)
```
